```python
m, P, t, k = symbols('m, P, t, k')
```

```python
MSE = 1/3 * ((m * 1945 + P * 560 + k - 4.3)**2 + \
             (m * 1495 + P * 340 + k - 4.9)**2 + \
             (m * 1570 + P * 343 + k - 5.2)**2)
MSE
```

```python
MSEm = diff(MSE, m)
MSEm
```

```python
MSEP = diff(MSE, P)
MSEP
```

```python
MSEk = diff(MSE, k)
MSEk
```

```python
nonlinsolve([MSEm, MSEP, MSEk], [m, P, k])
```
