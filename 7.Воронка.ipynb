```
import csv

funnel = {}
funnel_template = {'1_home_page': 0, '2_search_page': 0, '3_payment_page': 0, '4_payment_confirmation_page': 0}
file_path = 'click_stream3.csv'
with open(file_path, mode='r') as csv_file:  # открываем файл
    csv_reader = csv.DictReader(csv_file)  # читаем файл

    for row in csv_reader:  # перебираем по одной строчке нашего файла
        gender = list(row.items())[4][1]
        device = list(row.items())[3][1]
        event_date = list(row.items())[2][1][:-3]
        page = list(row.items())[1][1]

        if gender not in funnel:
            funnel[gender] = {}
        if device not in funnel[gender]:  # если нет
            funnel[gender][device] = {}  # создаем ключ, в значение копируем заготовку
        if event_date not in funnel[gender][device]:  # если нет
            funnel[gender][device][event_date] = funnel_template.copy()  # создаем ключ, в значение копируем заготовку
            if page == '1_home_page':
                funnel[gender][device][event_date]['1_home_page'] += 1
            elif page == '2_search_page':
                funnel[gender][device][event_date]['2_search_page'] += 1
            elif page == '3_payment_page':
                funnel[gender][device][event_date]['3_payment_page'] += 1
            else:
                funnel[gender][device][event_date]['4_payment_confirmation_page'] += 1
        else:
            if page == '1_home_page':
                funnel[gender][device][event_date]['1_home_page'] += 1
            elif page == '2_search_page':
                funnel[gender][device][event_date]['2_search_page'] += 1
            elif page == '3_payment_page':
                funnel[gender][device][event_date]['3_payment_page'] += 1
            else:
                funnel[gender][device][event_date]['4_payment_confirmation_page'] += 1

funnel
```
